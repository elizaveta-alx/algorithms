```
import networkx as nx 
import matplotlib.pyplot as plt 
import random 
import numpy as np
from timeit import default_timer as timer
```

# **Part 1**

```
#creating adjacency matrix
def create_adjacency_matrix(n, edges):
  adjacency_matrix = numpy.zeros((n, n))
  current_edge = 0
  while current_edge < edges:
    i = numpy.random.randint(0, n)
    j = numpy.random.randint(0, n)
    if i != j and adjacency_matrix[i][j] == 0:
      adjacency_matrix[i][j] = numpy.random.randint(0, 100)
      adjacency_matrix[j][i] = adjacency_matrix[i][j]
      current_edge += 1
  return adjacency_matrix
```

```
n = 100 #n - number of vertices
edges = 500 #edges - number of edges
adjMatrix = create_adjacency_matrix(n,edges)
```

```
#Visualizing the graph
G = nx.Graph(adjMatrix)
pos = nx.random_layout(G)
plt.figure(figsize=(20,10))
labels = {}
for k in range(n):
  labels[k] = "{0}".format(k)
nx.draw_networkx_labels(G, pos, labels, font_color="white")
nx.draw(G, pos)
```

```
source = numpy.random.randint(0, n)
route_dijkstra = nx.single_source_dijkstra_path(G, source)
print("The shortest paths finding by Dijkstra algorithm from {0} node:".format(source))
for i in route_dijkstra:
  print(i,": ", route_dijkstra[i])
```

```
n_experiments = 10
times = []
for i in range(n_experiments):
    start_time = timer()
    nx.single_source_dijkstra_path(G, source)
    times.append(timer() - start_time)
print("Average time execution of Dijkstra algorithm: ", np.array(times, dtype=np.float64).mean(axis=0))
```

```
route_bellman_ford = nx.single_source_bellman_ford_path(G, source)
print("The shortest paths finding by Bellman-Ford algorithm from {0} node:".format(source))
for j in route_bellman_ford:
  print(j,": ", route_bellman_ford[j])
```

```
n_experiments = 10
times = []
for i in range(n_experiments):
    start_time = timer()
    nx.single_source_bellman_ford_path(G, source)
    times.append(timer() - start_time)
print("Average time execution of Bellman Ford algorithm: ", np.array(times, dtype=np.float64).mean(axis=0))
```

```
print("The shortest path from 12 to 21 nodes finding by Bellman-Ford algorithm: ", route_bellman_ford[21])
print("The shortest path from 12 to 21 nodes finding by Dijkstra algorithm: ", route_dijkstra[21])
```

# **Part 2**

```
m=10
n=20
G = nx.grid_2d_graph(m, n)
barrier = 40
#creating the graph
for i in range(barrier):
    random_node = random.choice(list(G.nodes))
    G.remove_node(random_node)
```

```
#Visualizing the graph
plt.figure(figsize=(20,10))
pos = {(x,y):(y,-x) for x,y in G.nodes()}
nx.draw(G, pos,with_labels=True, font_color="white", node_color="purple", linewidths=25)
```

```
def a_algorithm():
    start = random.choice(list(G.nodes)) 
    finish = random.choice(list(G.nodes))
    print("Start", start, "finish", finish)
    shortest_path = nx.astar_path(G, start, finish)
    #Visualizing the graph
    plt.figure(figsize=(15,10))
    pos = {(x,y):(y,-x) for x,y in G.nodes()}

    nx.draw(G, pos,with_labels=True, font_color="white", node_color="purple", linewidths=25)
    nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='r', linewidths=25)
    path_edges = list(zip(shortest_path, shortest_path[1:]))
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='r', width=9, label='Astar path')
    print("Shortest path between nodes {0} and {1}: \n".format(start, finish),shortest_path)
```
